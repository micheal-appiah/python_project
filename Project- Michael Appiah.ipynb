### Replacing null values in each column with mean of each column and why?

```python
eachcolumn=['fixed acidity','citric acid','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','quality']
```

```python
for column in eachcolumn:
    data[column].fillna(data[column].mean(),inplace=True)
```

When data analysis involves missing values, a common technique is to substitute null values with the mean of the column. Removing rows with missing values can result in a loss of information and a decrease in sample size, which can negatively impact the accuracy and representativeness of the analysis. To retain the sample size and provide a reasonable estimate of the missing data, we can replace null values with the mean of the column. However, this approach assumes that the missing values are randomly distributed and the data is normally distributed. If the missing values are not randomly distributed or the data is not normally distributed, this approach may not be suitable for providing an accurate representation of the missing data. In such instances, alternative techniques such as interpolation, imputation, or eliminating rows/columns may be more appropriate.

### Z-Scores and Outliers
The z-score indicates how many standard deviations a particular data point is distant from the mean of the distribution.

An outlier is a data point that lies far beyond the range of other data points. A popular technique for detecting outliers involves setting a threshold of 3 standard deviations.





```python
for column in data.columns:
    if column != 'ID' :
    #Calculating the z-scores for each data point of each column
       z_scores = (data[column] - data[column].mean()) / data[column].std()
    #output below is all z_scores values for each column
       print(z_scores)

     #outliers are identified using a threshold of 3 standard deviations
       outliers = data[abs(z_scores) > 3]
```

```python
outliers
```

```python
for column in outliers.columns:
    if column != 'ID':
        plt.title(f'Outliers in {column}')
        plt.xlabel('Range')
        
        sns.scatterplot(outliers[column])
        plt.show()
```
